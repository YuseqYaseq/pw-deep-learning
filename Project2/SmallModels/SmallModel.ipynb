```python
import keras
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Flatten, MaxPooling2D, Conv2D, Dropout, BatchNormalization
from keras.activations import relu, softmax

import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print(x_train.shape, y_train.shape)
```

```python
x_trainN1 = (x_train - 127.5) / 255.0
y_trainN1  = keras.utils.to_categorical(y_train, num_classes=10)
```

```python
print(x_trainN1.mean(), y_trainN1.shape)
```

```python
print(x_trainN1.mean(), x_trainN1.var())
```

```python
def plot_history(history):
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.legend(['Train', 'Valid'], loc='upper right')
  plt.show()
```

```python
def getSmallNet():
  smallNet = keras.models.Sequential()
  smallNet.add(Conv2D(32, (5,5), strides=1, padding='same', activation=relu, input_shape=(32,32, 3)))
  smallNet.add(MaxPooling2D(pool_size=(2, 2)))
  smallNet.add(Conv2D(64, (3,3), strides=1, padding='same', activation=relu))
  smallNet.add(MaxPooling2D(pool_size=(2, 2)))
  smallNet.add(Conv2D(128, (3,3), strides=1, padding='same', activation=relu))
  smallNet.add(MaxPooling2D(pool_size=(2, 2)))
  smallNet.add(Flatten())
  smallNet.add(Dense(256, activation=relu))
  smallNet.add(Dense(64, activation=relu))
  smallNet.add(Dense(10, activation=softmax))
  return smallNet
```

```python
for lr in [0.5, 0.1, 0.05, 0.01, 0.005, 0.001]:
  smallNet = getSmallNet()
  opt = keras.optimizers.SGD(learning_rate=lr)
  smallNet.compile(loss='categorical_crossentropy',
                   optimizer=opt,
                   metrics=['accuracy'])

  history = smallNet.fit(x_trainN1, y_trainN1, batch_size=32, epochs=10, validation_split=0.1)
  plot_history(history)
```

```python
smallNet = getSmallNet()
opt = keras.optimizers.SGD(learning_rate=0.01)
smallNet.compile(loss='categorical_crossentropy',
                   optimizer=opt,
                   metrics=['accuracy'])

history = smallNet.fit(x_trainN1, y_trainN1, batch_size=32, epochs=100, validation_split=0.1)
plot_history(history)
```

```python
smallNet = getSmallNet()
opt = keras.optimizers.Adagrad(learning_rate=0.01)
smallNet.compile(loss='categorical_crossentropy',
                   optimizer=opt,
                   metrics=['accuracy'])

history = smallNet.fit(x_trainN1, y_trainN1, batch_size=32, epochs=100, validation_split=0.1)
plot_history(history)
```

```python
def getSmallNet2():
  smallNet = keras.models.Sequential()
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu, 
                      input_shape=(32,32, 3)
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Conv2D(128, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu
                      ))
  smallNet.add(MaxPooling2D(pool_size=(4, 4), 
                            strides=(1,1)))
  smallNet.add(Flatten())
  smallNet.add(Dense(32, 
                     activation=relu,
                    ))
  smallNet.add(Dense(10, activation=softmax))
  return smallNet
```

```python
smallNet = getSmallNet2()
smallNet.summary()
```

```python
smallNet = getSmallNet2()
opt = keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
smallNet.compile(loss='categorical_crossentropy',
                   optimizer=opt,
                   metrics=['accuracy'])

history = smallNet.fit(x_trainN1, y_trainN1, batch_size=256, epochs=300, validation_split=0.1)
plot_history(history)
```

```python
def getSmallNet3():
  smallNet = keras.models.Sequential()
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      input_shape=(32,32, 3)
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Dropout(rate=0.3))

  smallNet.add(BatchNormalization())
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Dropout(rate=0.3))

  smallNet.add(BatchNormalization())

  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D(pool_size=(4, 4), 
                            strides=(1,1)))
  smallNet.add(Dropout(rate=0.3))
  smallNet.add(BatchNormalization())

  smallNet.add(Flatten())

  smallNet.add(Dense(128, 
                     activation=relu,
                    ))
  smallNet.add(BatchNormalization())
  smallNet.add(Dropout(rate=0.3))
  smallNet.add(Dense(10, 
                    activation=softmax))
  return smallNet
```

```python
def getSmallNet4():
  smallNet = keras.models.Sequential()
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      input_shape=(32,32, 3)
                      ))
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='same', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Dropout(rate=0.1))

  smallNet.add(BatchNormalization())
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Dropout(rate=0.1))

  smallNet.add(BatchNormalization())

  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D(pool_size=(2, 2), 
                            strides=(2,2)))
  smallNet.add(Dropout(rate=0.1))

  smallNet.add(BatchNormalization())

  smallNet.add(Conv2D(128, 
                      (3,3), 
                      strides=1, 
                      padding='same', 
                      activation=relu,
                      ))
  smallNet.add(Dropout(rate=0.2))
  smallNet.add(BatchNormalization())

  smallNet.add(Flatten())

  smallNet.add(Dense(256, 
                     activation=relu,
                    ))
  smallNet.add(BatchNormalization())
  smallNet.add(Dropout(rate=0.5))
  smallNet.add(Dense(256, 
                     activation=relu,
                    ))
  smallNet.add(Dropout(rate=0.5))
  smallNet.add(Dense(10, 
                    activation=softmax))
  return smallNet
```

```python
smallNet = getSmallNet4()
smallNet.summary()
```

```python
def getSmallNet5():
  smallNet = keras.models.Sequential()
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      input_shape=(32,32, 3),
                      ))
  smallNet.add(Dropout(rate=0.2))
  smallNet.add(BatchNormalization())
  smallNet.add(Conv2D(64, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(Dropout(rate=0.2))
  smallNet.add(BatchNormalization())
  smallNet.add(Conv2D(32, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(Dropout(rate=0.2))

  smallNet.add(BatchNormalization())
  smallNet.add(Conv2D(32, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(Dropout(rate=0.2))
  smallNet.add(BatchNormalization())

  smallNet.add(Conv2D(32, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(Dropout(rate=0.2))
  smallNet.add(BatchNormalization())
  smallNet.add(Conv2D(32, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D((2,2)))
  smallNet.add(Dropout(rate=0.2))

  smallNet.add(BatchNormalization())

  smallNet.add(Conv2D(32, 
                      (3,3), 
                      strides=1, 
                      padding='valid', 
                      activation=relu,
                      ))
  smallNet.add(MaxPooling2D((2,2)))
  smallNet.add(Dropout(rate=0.2))
  smallNet.add(BatchNormalization())

  smallNet.add(Flatten())

  smallNet.add(Dense(256, 
                     activation=relu,
                    ))
  smallNet.add(BatchNormalization())
  smallNet.add(Dropout(rate=0.5))
  smallNet.add(Dense(128, 
                     activation=relu,
                    ))
  smallNet.add(Dense(10, 
                    activation=softmax,
                     ))
  return smallNet
```

```python
smallNet = getSmallNet5()
smallNet.summary()
```

```python
smallNet = getSmallNet5()
opt = keras.optimizers.Adam(learning_rate=0.005, beta_1=0.9, beta_2=0.999)
#opt = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
smallNet.compile(loss='categorical_crossentropy',
                 optimizer=opt,
                 metrics=['accuracy'])
history = smallNet.fit(x_trainN1, y_trainN1, batch_size=256, epochs=300, validation_split=0.1)
plot_history(history)
```
